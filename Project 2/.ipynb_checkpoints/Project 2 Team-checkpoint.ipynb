```python
#
# Recode variables
#-----------

# Diabetes
# ----------
# keep only useful info about our target feature, i.e., where diabetes_number not = 'Suppressed'
# note that the inside reference to the diabetes df identifies the column, and the outer calls specific rows according to a condition 
diabetes = diabetes[diabetes['Diabetes_Number']!="Suppressed"] 

# use the astype method on Diabetes_Number to convert it to an integer...if you are not sure, what does the astype() documentation tell you are possible arguments? 
diabetes['Diabetes_Number'] = diabetes['Diabetes_Number'].astype(np.int64) 

# Obesity
diabetes = diabetes[diabetes['Obesity_Number']!="No Data"] 
diabetes['Obesity_Number'] = diabetes['Obesity_Number'].astype(np.int64) 


# Physical Inactivity
diabetes = diabetes[diabetes['Physical_Inactivity_Number']!="No Data"] 
diabetes['Physical_Inactivity_Number'] = diabetes['Physical_Inactivity_Number'].astype(np.int64) 

# Some final changes .
# ----------
# 65+ sex ratio had one "-" in it so let's drop that row first
diabetes = diabetes[diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)']!= "-"]

# change to numeric from string, since it originally included the "-", which made it a string
# you'll have to decide whether to make it integer or float 
diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)'] = diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)'].astype(np.float64)
```

#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

### 3.2 Partition Data, Feature Selection, and Standardization

Now, we will partition our data to prepare it for the training process. Ultimately we want to use a 60% train—20% validation—20% test in this case. More data in the training set lowers bias, but then increases variance in the validation/test sets. Balancing between bias and variance with choice of these set sizes is important as we want to ensure that there is enough data to train on to get good predictions, but also want to make sure our hold-out sets are representative enough.

Work through partitioning the data into the test/train/validation sets in the chunks below. Be sure to that if you are using Ridge or LASSO, you Standardize the data. Where you do this in the workflow matters so be clear about where you are doing this and why. 

**Workflow discussion:** The training set split is used to initially train the selected models on a subset of the data – we use 60% of the data here to give the training models a decent amount of information while leaving enough for both the validation and test sets (20% each). The validation set allows us to potentially then see how the training models perform on a subset of the data that the training model has not seen yet – here is where we can also test model performance and do any necessary hyperparameter tuning to all our models and further decide upon a “best” model. Finally, the test set is used to again assess how well the training model, after being hyperparameter tuned, performs on a final subset of the full data that the training model has yet to see. 

Standardization is required for LASSO and ridge models since those models require that all betas have a common metric. Machine learning models that train and predict on test/train splits require feature standardization AFTER splitting the data to avoid data leakage, which occurs when test or validation model information enters the training data. This can lead to biased prediction quality in the test data if there is test data information in the training process. 


```python
#
# Partition data
#-----------

# import library
from sklearn.model_selection import train_test_split

# create y dataframe 
y = diabetes_clean['Diabetes_Number']

# create X dataframe (include everything except "Diabetes_Number", our target, 
# and "race alone or in combination with one or more other races_total population")
X = diabetes_clean.drop(['Diabetes_Number', 'race alone or in combination with one or more other races_total population','CountyFIPS'
                                                 ], axis=1)
```

```python
#
# Training/test split
#-----------

# set the random seed
np.random.seed(10)


# split the data so that it returns 4 values: X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X,                 # specify training dataset
                                                    y,                 # specify test dataset
                                                    train_size= 0.80,      # specify proportional split for training
                                                    test_size= 0.20)       # specify proportional split for test
```

```python
#
# Validation split
#-----------

# train_test_split returns 4 values: X_train, X_test, y_train, y_test, so how do we create a 60-20-20 train-validate-test split? 
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, train_size= 0.75, test_size= 0.25)    

print(f"Train: {len(X_train)} | Validate: {len(X_validate)} | Test: {len(X_test)}")
```

```python
#
# Standardization
#-----------
# Given that we want to only standardize non-dichotomous variables, we need to find a 
# solution that will loop over only the columns we want to standardize. The code below
# identifies all non-dichotomous variables in our dataset and only standardizes those.

# load library and create instance of Standard Scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


# identify non-dichotomous columns we want to transform
columns = list(X_test.select_dtypes(include=['number']).loc[:, X_test.nunique() > 2])

# use loop to transform training data for only columns we want to transform
for column in columns:
    X_train[column] = scaler.fit_transform(X_train[column].values.reshape(-1, 1)).flatten()

# use loop to transform validation data for only columns we want to transform
for column in columns:
    X_validate[column] = scaler.fit_transform(X_validate[column].values.reshape(-1, 1)).flatten()

# use loop to transform test data for only columns we want to transform
for column in columns:
    X_test[column] = scaler.fit_transform(X_test[column].values.reshape(-1, 1)).flatten()
```

### 4.1 Describe models

Detail the basic logic and assumptions underlying each model, its pros/cons, and why it is a plausible choice for this problem.




**MODEL DESCRIPTION(S):** ...

#### **OLS:**

***Basic logic:*** estimates the linear relationship between independent and dependent variables that minimizes the sum of squared errors. 

***Assumptions:*** there is a linear relationship between independent and dependent variables, the independent variables are uncorrelated with the error term, and the error terms are homoscedastic.  
***Pros:*** Coefficients are easy to interpret and it is rare to hyperparameter tune – more of a question of understanding the particular research question and specifics of context that generate model-building processes.

***Cons:*** Estimates can be easily biased due to unmeasured confounding, and when there are too many predictors, the model can overfit the estimates by forcing too many linear relationships. 

#### **LASSO:**

***Basic logic:*** LASSO is a regression with an L1 error term (absolute value of beta) that penalizes mean loss to standardize the X’s (aka the features).

***Assumptions:*** the model is complex and requires a penalty, X’s (features) must be standardized, all other OLS assumptions. 

***Pros:*** reduces overfitting with shrinkage and regularization. 

***Cons:*** LASSO throws out collinear variables, so you have to do some theoretical work to decide which stays and which goes, which is something that the model would not know.

#### **KNN:**

***Basic logic:*** K Nearest Neighbors is a non-parametric model that looks at the spatial (Euclidean) distance points have from one another.


***Assumptions:*** distance can logically capture similarity, nearby points are assumed to have similar responses, observations are independent, sample size is big enough; also, features must be standardized since you are comparing them.


***Pros:*** flexible and intuitive since you look at nearest data points and have a non-parametric (aka, no underlying distribution in the data), and it can be used for classification and regression. 


***Cons:*** it’s sensitive to irrelevant features or outliers in data, it is not as interpretable since it doesn’t give coefficient estimates like an OLS or LASSO model would, and it can be computationally expensive.

**What is the advantage of using both validation and test sets in the social sciences and public policy?:**

Using both a validation and test set ensures that a predictive or policy evaluation model is optimized and has a reliable, unbiased estimate of its training performance on new, unseen data. The validation set is used during the model development process to tune its hyperparameters and assess its performance iteratively. The test set is a final, untouched portion of the data that is used after all model training and tuning is complete. Since the model has never been exposed to the test set, the performance metrics derived from it provided the most unbiased and realistic estimate of how the final model will perform when deployed in a real-world social or policy environment with truly new data.

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

Bias is an error arising from the residuals, and having a high bias tends to underfit the dataset. Variance is the error from noise that is captured in the output, which tends to be too complex and overfit the dataset. Thus, when bias increases, variance decreases, and visa-versa. The tradeoff between bias and variance within machine learning is trying to fit a model that is complex enough to predict well on when new data is introduced, while also accurately being trained on a dataset and learning its patterns. Data scientists thus must make modeling decisions on how to implement models and their adjustments to balance both bias and variance. This tradeoff is relevant to machine learning problems like this one because of the number of features our models are trained on and how many counties are present in the dataset. Counties within the US surely have various similarities and differences as indicated by the various features in our dataset. A model with high variance might be too complex and noisy, and poorly capture the generalizations that we are seeing between counties to make a policy decision at a population health level. 


2. Define overfitting, and why it matters for machine learning. How can we address it?

 Overfitting is when the model has learned the complexities of the training models too well, capturing both the underlying patterns of the features as well as predicting random noise or outliers into the outcome. Thus, a model that is overfit will perform poorly on a dataset that it has not seen yet. This matters because machine learning is about prediction: the goal is to train models such that the models are able to make accurate predictions and uncover patterns when introduced to a new sample that the model was not trained on. We can use some machine learning techniques to address overfitting. LASSO and Ridge add penalties on coefficients for models that are too complex, preventing overfitting. Cross-Validation also assesses a model’s prediction performance by running the model through multiple combinations of training and test sets of the data it’s being given, to provide a more reliable and generalizable estimate.
